```python
#Mean and Standard Deviation
import math
def mean(numbers):
    return sum(numbers[0:296])/float(296),sum(numbers[296:])/float(357)

def stdev(numbers):
    avgp,avgn = mean(numbers)
    variancep = sum([pow(x-avgp,2) for x in numbers[0:296]])/float(295)
    variancen = sum([pow(x-avgn,2) for x in numbers[296:]])/float(356)
    return math.sqrt(variancep),math.sqrt(variancen)

#calculate probabilities
def calculateProbability(x, meanp,meann, stdevp,stdevn):
    exponentp = math.exp(-(math.pow(x-meanp,2)/(2*math.pow(stdevp,2))))
    exponentn = math.exp(-(math.pow(x-meann,2)/(2*math.pow(stdevn,2))))
    return (1 / (math.sqrt(2*math.pi) * stdevp)) * exponentp,(1 / (math.sqrt(2*math.pi) * stdevn)) * exponentn
#likelihood for continuous values
def contlikelihood(x,numbers):
    pmean,nmean=mean(numbers)
    pstd,nstd=stdev(numbers)
    pc,nc=calculateProbability(x,pmean,nmean,pstd,nstd)
    return pc,nc
#contlikelihood(40.89,c2)
```

```python
#p(class/feature)=count(feature in class)/count(class) likelihood for discrete
def likelihood(x,tdata):
    #stdata=set(tdata)
    #datadict={val:key+1 for key,val in enumerate(list(stdata))}
    #print(x,datadict)
    #tval=datadict[x]
    tdata=rearrange(tdata,index)
    pfeaturecount=len([i for i in tdata[0:296] if x==i])
    #print(pfeaturecount)
    probpos=(pfeaturecount+1)/296.0
    nfeaturecount=len([i for i in tdata[296:] if x==i])
    #print(nfeaturecount)
    negpos=(nfeaturecount+1)/357.0
    return probpos,negpos
x,y=likelihood("m",c6)
x,y
```

```python
def NaiveBayesJ(a,b,c,d,e,f,g,h,i,j,k,l,m,n,o):
    a1,a2=likelihood(a,c1)
    b1,b2=contlikelihood(b,c2)
    cp,cn=contlikelihood(c,c3)
    d1,d2=likelihood(d,c4)
    e1,e2=likelihood(e,c5)
    f1,f2=likelihood(f,c6)
    g1,g2=likelihood(g,c7)
    h1,h2=contlikelihood(h,c8)
    i1,i2=likelihood(i,c9)
    j1,j2=likelihood(j,c10)
    k1,k2=contlikelihood(k,c11)
    l1,l2=likelihood(l,c12)
    m1,m2=likelihood(m,c13)
    n1,n2=contlikelihood(n,c14)
    o1,o2=contlikelihood(o,c15)
    ppos=296.0/653.0
    pneg=357.0/653.0
    postive=a1*b1*cp*d1*e1*f1*g1*h1*i1*j1*k1*l1*m1*n1*o1*ppos
    negative=a2*b2*cn*d2*e2*f2*g2*h2*i2*j2*k2*l2*m2*n2*o2*pneg
    return postive,negative   
```
